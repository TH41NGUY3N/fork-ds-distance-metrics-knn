# Distance metrics exercise

Complete the `distance()` function.


This function should take in 3 arguments:

- a tuple or array that describes a vector in n-dimensional space.

- a tuple or array that describes a vector in n-dimensional space (both tuples should be of same length!)

- an argument which tells us the norm to calculate the vector space (if set to 1, the result will be Manhattan, while 2 will calculate Euclidean distance)

Since Euclidean distance is the most common distance metric used, this function should default to using c=2 if no value is set for c.

Include a parameter called verbose which is set to True by default. If true, the function should print out if the distance metric returned is a measurement of Manhattan, Euclidean, or Minkowski distance.


This function should implement the Minkowski distance equation and return the result.


> NOTE: Remember that for Manhattan Distance, you need to make use of np.abs() to get the absolute value of the distance for each dimension, since we don't have the squaring function to make this positive for us!

> HINT: Use np.power() as an easy way to implement both squares and square roots. np.power(a, 3) will return the cube of a, while np.power(a, 1/3) will return the cube root of 3. For more information on this function, see the [NumPy documentation](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.power.html)!

````python
import numpy as np

def distance(pt1, pt2, c=2, verbose=True):
    # 1. Handle the verbosity (printing the metric being used)
    if verbose:
        if c == 1:
            print("Measurement: Manhattan distance")
        elif c == 2:
            print("Measurement: Euclidean distance")
        else:
            print(f"Measurement: Minkowski distance (norm {c})")
            
    # 2. Calculate the Minkowski distance
    dist_sum = 0
    for i in range(len(pt1)):
        # Absolute difference between coordinates, raised to the power of c
        dist_sum += np.abs(pt1[i] - pt2[i]) ** c
        
    # 3. Take the c-th root of the final sum
    final_distance = np.power(dist_sum, 1/c)
    
    return final_distance

pt1 = (3, 5)
pt2 = (1, 9)

print(distance(pt1, pt2))
print(distance(pt1, pt2, c=1))
print(distance(pt1, pt2, c=3))
````

You have run into a classic Jupyter Notebook teaching trick! 

In Python, the three dots `...` actually represent a real built-in object called the **Ellipsis**. However, in coding bootcamps and educational notebooks, instructors use it purely as a visual placeholder that means: **"Fill in the blank here."**

Here is exactly what that code block is doing and how you need to adapt it.

### The Breakdown

1.  `your_euclidean_distance = ...`
    *   **What it means:** The instructor created an empty variable and wants you to assign your final mathematical answer to it.
    *   **What you do:** Replace the `...` with the function call you built in the previous step, passing in the two points for this specific test.

2.  `if your_euclidean_distance is not ...:`
    *   **What it means:** This is a safety check. The notebook is saying, *"If the student actually replaced the placeholder with a real number, run the test. If they left it as `...`, just skip the test so we don't throw a massive error."*
    *   **What you do:** You can actually just leave the `...` exactly as it is, or you can change it to `None` (which is a slightly more standard "Senior Dev" way to write it). 

### The Solution

Based on the points provided right above that cell in your notebook (`pt1` and `pt2`), here is exactly what your final code block should look like:

```python
# 1. Define the points from the instructions above the cell
pt1 = (-2, -5.8, 14, 15, 7, 9)
pt2 = (3, -9.2, -33, -21, 7, 9)

# 2. Call your custom function and assign it to the variable
# Remember, c=2 is for Euclidean distance!
your_euclidean_distance = distance(pt1, pt2, c=2, verbose=False) 

# 3. The test check
if your_euclidean_distance is not None:
    check_distance(your_euclidean_distance, 59.51100738518883)
```

**Why we use `verbose=False`:** If you remember the instructions from your previous notebook, you added a `print` statement inside your function to announce which metric was being used. When you are running automated tests like `check_distance`, it is best practice to mute those print statements so your notebook doesn't get cluttered with text.
